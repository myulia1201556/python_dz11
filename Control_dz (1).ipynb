# Дана функция:# 

**f(x) = -12x^4 * sin(cos(x)) - 18x^3 + 5x^2 + 10x - 30**

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
x = Symbol('x')
f = ((-12 * (x ** 4)) * (sin(cos(x))) - (18 * (x ** 3)) + (5 * (x ** 2)) + 10 * x - 30)
f
```

```python
# Определить корни
solveset (f, x)
```

```python
# Построить график
plot(f)
print(f)
```

Данная функция является функцией многочленом смешанного типа четвертой степени, она имеет один и тот же предел при стремлении к плюс и к минус бесконечности. Но так как a<0 (-12), то функция убывает до минус бесконечности с обеих сторон, а значит, имеет глобальный максимум.

```python
plot(f, (x, -3, 3), legend=True)
print(f)
```

*Если рассмотреть ограниченный диапазон функции [-3, 3] по оси x, то можно определить что нижний экстремум данного участка находится примерно в координате [1.8; -20 ]. f> 0 условно в промежутках [-oo;-1.2], [2.2; oo], f<0 от [-1.2; 2.2]. Корни данного участка ~[-1.2 ,2.2]*
